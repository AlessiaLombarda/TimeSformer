**Preprocessing of the whole dataset**

```python
os.mkdir("UCF-11_preprocessed")
```

```python
for dir in os.listdir("/content/action_youtube_naudio"):
  if dir!="readme.txt":
    os.mkdir("UCF-11_preprocessed/"+dir)
    for dir2 in os.listdir("/content/action_youtube_naudio/"+dir):
      if dir2!="Annotation":
        for file in os.listdir("/content/action_youtube_naudio/"+dir+"/"+dir2):
          if fnmatch.fnmatch(file, '*.avi'):
            clip = VideoFileClip("/content/action_youtube_naudio/"+dir+"/"+dir2+"/"+file).resize((224,224))
            sample = np.arange(0, clip.duration, clip.duration/8)
            sample_frame = [clip.get_frame(i) for i in sample]
            clip.close()
            clip = ImageSequenceClip(sample_frame, fps = 8)
            clip.write_videofile("UCF-11_preprocessed/"+dir+"/"+file, codec='rawvideo')
            clip.close() 
```

**Create csv files, 50% training, 20% validation, 30% test**

```python
f_train = open("/content/UCF-11_preprocessed/train.csv", "a")
f_val = open("/content/UCF-11_preprocessed/val.csv", "a")
f_test = open("/content/UCF-11_preprocessed/test.csv", "a")
f_label = open("/content/UCF-11_preprocessed/label.csv", "a")
i=0

for dir in os.listdir("/content/UCF-11_preprocessed"):
  if(dir!="test.csv" and dir!="train.csv" and dir!="val.csv"and dir!="label.csv" and dir!=".ipynb_checkpoints"):
    f_label.write(dir+" "+str(i)+"\n")
    for x in os.listdir("/content/UCF-11_preprocessed/"+dir):
       rand = rnd.random()
       if(rand < 0.2):
         f_val.write("/home/studenti/lombarda/UCF-11_preprocessed/"+dir+"/"+x+" "+str(i)+"\n")
       elif (rand>=0.2 and rand<0.5):  
         f_test.write("/home/studenti/lombarda/UCF-11_preprocessed/"+dir+"/"+x+" "+str(i)+"\n")
       else:
         f_train.write("/home/studenti/lombarda/UCF-11_preprocessed/"+dir+"/"+x+" "+str(i)+"\n")  
    i+=1

f_test.close()  
f_val.close()
f_train.close()
f_label.close()
```

**Check label distribution in train, test and validation set**

```python
def get_labels(file):
  f = open(file, "r")
  label = np.zeros(11)
  for x in f:
    l = int(x.split(" ")[1])
    label[l-1] += 1
  return label  
```

```python
train_label = get_labels("/content/UCF-11_preprocessed/train.csv")
val_label = get_labels("/content/UCF-11_preprocessed/val.csv")
test_label = get_labels("/content/UCF-11_preprocessed/test.csv")

plt.figure(figsize=[25, 6])
plt.subplot(1,3,1)
plt.bar(np.arange(11), train_label)
plt.subplot(1,3,2)
plt.bar(np.arange(11), val_label)
plt.subplot(1,3,3)
plt.bar(np.arange(11), test_label)

plt.show()
```
