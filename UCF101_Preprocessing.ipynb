**Preprocessing of the whole dataset**

```python
os.mkdir("UCF-101_preprocessed")
```

```python
dirs = os.listdir("/content/UCF-101")
len = len(dirs)
```

```python
for dir in dirs:
  os.mkdir("UCF-101_preprocessed/"+dir)
  for files in os.listdir("/content/UCF-101/"+dir):
    clip = VideoFileClip("/content/UCF-101/"+dir+"/"+files).resize((224,224))
    sample = np.arange(0, clip.duration, clip.duration/8)
    sample_frame = [clip.get_frame(i) for i in sample]
    clip.close()
    clip = ImageSequenceClip(sample_frame, fps = 8)
    clip.write_videofile("UCF-101_preprocessed/"+dir+"/"+files, codec='rawvideo')
    clip.close()
    os.remove("/content/UCF-101/"+dir+"/"+files)
  os.rmdir("/content/UCF-101/"+dir)    
```

```python
os.rmdir("UCF-101")
```

```python
f1 = open("/content/dataset.csv", 'a')
i = 0
for dir in os.listdir("/content/UCF-101_preprocessed"):
    for files in os.listdir("/content/UCF-101_preprocessed/"+dir):
      f1.write("/content/UCF-101_preprocessed/"+dir+"/"+files+" "+str(i)+"\n")
    i+=1

f1.close()
shutil.copyfile("/content/dataset.csv", "/content/UCF-101_preprocessed/dataset.csv")
```
